#### Using iris dataset

```julia
using Gadfly
using RDatasets

iris = dataset("datasets", "iris")
```

#### Plotting Coherent 

```julia
plot(iris, x=:SepalLength, y=:SepalWidth,
         Geom.point, Geom.line)
```

#### Plotting Coherent adding colors

```julia
plot(iris, x=:SepalLength, y=:SepalWidth, color=:Species,
         Geom.point)
```
